```python
# List of the columns used in the rCSI formula
rcsi_cols = [
    'CSI_ReliedOnLess',
    'CSI_BorrowedFood',
    'CSI_ReducedNoOfMeals',
    'CSI_ReducedPortionMealSize',
    'CSI_QuantityForAdult'
]

# Replace missing (NaN) values with 0
df1[rcsi_cols] = df1[rcsi_cols].fillna(0)

# Calculate rCSI and store in a new column
df1['rCSI'] = (
    df1['CSI_ReliedOnLess'] * 2 +
    df1['CSI_BorrowedFood'] * 2 +
    df1['CSI_ReducedNoOfMeals'] * 1 +
    df1['CSI_ReducedPortionMealSize'] * 1 +
    df1['CSI_QuantityForAdult'] * 3
)
df1.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Fill missing with 0
rcsi_cols = [
    'CSI_ReliedOnLess',
    'CSI_BorrowedFood',
    'CSI_ReducedNoOfMeals',
    'CSI_ReducedPortionMealSize',
    'CSI_QuantityForAdult'
]
df1[rcsi_cols] = df1[rcsi_cols].fillna(0)

# Step 2: Calculate rCSI
df1['rCSI'] = (
    df1['CSI_ReliedOnLess'] * 2 +
    df1['CSI_BorrowedFood'] * 2 +
    df1['CSI_ReducedNoOfMeals'] * 1 +
    df1['CSI_ReducedPortionMealSize'] * 1 +
    df1['CSI_QuantityForAdult'] * 3
)

# Step 3: Categorize rCSI
bins = [-1, 3, 18, float('inf')]
labels = ['None', 'Stressed', 'Crisis']
df1['rCSI_Category'] = pd.cut(df1['rCSI'], bins=bins, labels=labels)

# Step 4: Calculate percentages
category_counts = df1.groupby(['County', 'rCSI_Category']).size().reset_index(name='Count')
total_counts = df1.groupby('County').size().reset_index(name='Total')
category_percent = category_counts.merge(total_counts, on='County')
category_percent['Percentage'] = (category_percent['Count'] / category_percent['Total']) * 100

# Step 5: Pivot for plotting
plot_df = category_percent.pivot(index='County', columns='rCSI_Category', values='Percentage').fillna(0)
plot_df = plot_df[['None', 'Stressed', 'Crisis']]  # Ensure correct order

# Step 6: Plot with custom colors
colors = {'None': 'green', 'Stressed': 'yellow', 'Crisis': 'red'}
ax = plot_df.plot(kind='bar', stacked=True, figsize=(12, 6), color=[colors[col] for col in plot_df.columns])

# Step 7: Add percentage labels on each bar segment
for idx, county in enumerate(plot_df.index):
    bottom = 0
    for cat in plot_df.columns:
        value = plot_df.loc[county, cat]
        if value > 0:
            ax.text(
                idx,
                bottom + value / 2,
                f'{value:.1f}%',
                ha='center',
                va='center',
                fontsize=9,
                color='black'
            )
        bottom += value

# Final plot settings
plt.ylabel("Percentage of Households")
plt.title("rCSI Categories by County")
plt.xticks(rotation=45)
plt.legend(title='rCSI Category')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Define FCS columns and fill NaN with 0
fcs_cols = [
    'HFC_GrainDays',
    'HFC_PulsesNutsDays',
    'HFC_GreenLeafyDays',
    'HFC_OrangeFruitsDays',
    'HFC_MeatDays',
    'HFC_MilkDays',
    'HFC_SugarDays',
    'HFC_OilDays'
]
df1[fcs_cols] = df1[fcs_cols].fillna(0)

# Step 2: Calculate FCS score
df1['FCS'] = (
    df1['HFC_GrainDays'] * 2 +
    df1['HFC_PulsesNutsDays'] * 3 +
    df1['HFC_GreenLeafyDays'] * 1 +
    df1['HFC_OrangeFruitsDays'] * 1 +
    df1['HFC_MeatDays'] * 4 +
    df1['HFC_MilkDays'] * 4 +
    df1['HFC_SugarDays'] * 0.5 +
    df1['HFC_OilDays'] * 0.5
)

# Step 3: Categorize FCS
fcs_bins = [-1, 21, 35, float('inf')]
fcs_labels = ['Poor', 'Borderline', 'Acceptable']
df1['FCS_Category'] = pd.cut(df1['FCS'], bins=fcs_bins, labels=fcs_labels)

# Step 4: Calculate % by County
fcs_counts = df1.groupby(['County', 'FCS_Category']).size().reset_index(name='Count')
total_per_county = df1.groupby('County').size().reset_index(name='Total')
fcs_percent = fcs_counts.merge(total_per_county, on='County')
fcs_percent['Percentage'] = (fcs_percent['Count'] / fcs_percent['Total']) * 100

# Step 5: Pivot for plotting
plot_df = fcs_percent.pivot(index='County', columns='FCS_Category', values='Percentage').fillna(0)
plot_df = plot_df[['Acceptable', 'Borderline', 'Poor']]  # Ensure proper order

# Step 6: Plot with custom colors
fcs_colors = {'Acceptable': 'green', 'Borderline': 'yellow', 'Poor': 'red'}
ax = plot_df.plot(kind='bar', stacked=True, figsize=(12, 6),
                  color=[fcs_colors[col] for col in plot_df.columns])

# Step 7: Add percentage labels on each bar segment
for idx, county in enumerate(plot_df.index):
    bottom = 0
    for cat in plot_df.columns:
        value = plot_df.loc[county, cat]
        if value > 0:
            ax.text(
                idx,
                bottom + value / 2,
                f'{value:.1f}%',
                ha='center',
                va='center',
                fontsize=9,
                color='black'
            )
        bottom += value

# Final plot settings
plt.ylabel("Percentage of Households")
plt.title("FCS Categories by County")
plt.xticks(rotation=45)
plt.legend(title='FCS Category')
plt.tight_layout()
plt.show()
```

```python
# Define the columns you want to keep
columns_to_keep = ['County', 'SubCounty', 'Ward', 'LivelihoodZone', 'Month', 'MUAC']

# Create df3 with only those columns
df3 = df2[columns_to_keep]
df3
```

```python
import numpy as np

# Define conditions for MUAC classification
conditions = [
    df3['MUAC'] < 115,
    (df3['MUAC'] >= 115) & (df3['MUAC'] < 125),
    (df3['MUAC'] >= 125) & (df3['MUAC'] < 135),
    df3['MUAC'] >= 135
]

# Define corresponding categories
categories = ['Severe', 'Moderate', 'At Risk', 'Healthy']

# Create new column
df3['MUAC_Props'] = np.select(conditions, categories, default='Unknown')
df3
```

```python
import matplotlib.pyplot as plt

# Step: Group and calculate percentages
muac_counts = df3.groupby(['County', 'MUAC_Props']).size().reset_index(name='Count')
total_counts = df3.groupby('County').size().reset_index(name='Total')
muac_percent = muac_counts.merge(total_counts, on='County')
muac_percent['Percentage'] = (muac_percent['Count'] / muac_percent['Total']) * 100

# Pivot for plotting
plot_df = muac_percent.pivot(index='County', columns='MUAC_Props', values='Percentage').fillna(0)

# Optional: sort columns for consistent color order
category_order = ['Severe', 'Moderate', 'At Risk', 'Healthy']
plot_df = plot_df[category_order]

# Custom colors
colors = {
    'Severe': 'red',
    'Moderate': 'black',
    'At Risk': 'yellow',
    'Healthy': 'green'
}

# Plot
ax = plot_df.plot(kind='bar', stacked=True, figsize=(12, 6),
                  color=[colors[col] for col in plot_df.columns])

plt.ylabel("Percentage of Children")
plt.title("MUAC Classification by County")
plt.xticks(rotation=45)
plt.legend(title='MUAC Category')
plt.tight_layout()
plt.show()
```
